## Read the Data

```python
from scipy.io import loadmat

def read_HSI():
  X = loadmat('Data_satellite/Indian_pines_corrected.mat')['indian_pines_corrected']
  y = loadmat('Data_satellite/Indian_pines_gt.mat')['indian_pines_gt']
  print(f"X shape: {X.shape}\ny shape: {y.shape}")
  return X, y

X, y = read_HSI()
```

## Convert the dataset into csv 

```python
import pandas as pd
import numpy as np

def extract_pixels(X, y):
  q = X.reshape(-1, X.shape[2])
  df = pd.DataFrame(data = q)
  df = pd.concat([df, pd.DataFrame(data = y.ravel())], axis=1)
  df.columns= [f'band{i}' for i in range(1, 1+X.shape[2])]+['class']
  df.to_csv('Dataset.csv')
  return df
  
df = extract_pixels(X, y)
```

## Principal Component Analysis(PCA)


```python
from sklearn.decomposition import PCA
X_reshaped = X.reshape(-1, X.shape[2])

# Standardize the features 
X_standardized = (X_reshaped - X_reshaped.mean(axis=0)) / X_reshaped.std(axis=0)

# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_standardized)

# DataFrame for principal components and labels
columns = [f'PC_{i+1}' for i in range(X_pca.shape[1])]
df_pca = pd.DataFrame(data=X_pca, columns=columns)
df_pca['label'] = y.flatten()

# Explained Variance Ratio
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# Plot explained variance ratio
plt.figure(figsize=(12, 6))
plt.plot(explained_variance_ratio, marker='o')
plt.title('Explained Variance Ratio of Principal Components')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.grid(True)
plt.show()

# Plot cumulative explained variance
plt.figure(figsize=(12, 6))
plt.plot(cumulative_explained_variance, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance of Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(12, 10))

df_filtered = df_pca[df_pca['label'] != 0]

fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(df_filtered['PC_1'], df_filtered['PC_2'], df_filtered['PC_3'], c=df_filtered['label'], cmap='viridis', alpha=0.3)

ax.set_title('3D Scatter Plot of PC1, PC2, and PC3 (Excluding Label 0)')
ax.set_xlabel('Principal Component 1 (PC1)')
ax.set_ylabel('Principal Component 2 (PC2)')
ax.set_zlabel('Principal Component 3 (PC3)')
ax.legend(*scatter.legend_elements(), title='Labels')

plt.show()
```

```python
from numpy.random import rand
from sklearn.decomposition import PCA

# PCA transform
data = rand(100, 10)
n_dim = data.shape[1]
pca = PCA(n_dim)
transformed = pca.fit_transform(data)

# Inverse PCA
def inverse_pca(pca_data, pca, remove_n):
    transformed = pca_data.copy()
    transformed[:, -remove_n:] = 0 
    return pca.inverse_transform(transformed)

new_data = inverse_pca(transformed, pca, 8)

common_vmin = min(np.min(new_data), np.min(data))
common_vmax = max(np.max(new_data), np.max(data))

fig = plt.figure(figsize = (6, 6))
fig.add_subplot(2,3, 1)
plt.imshow(new_data, vmin=common_vmin, vmax=common_vmax)
plt.colorbar()
fig.add_subplot(2,3, 2)
plt.imshow(data, vmin=common_vmin, vmax=common_vmax)
plt.colorbar()
plt.show()
```

## Classification Map

```python
#Use the function below to produce the classification map


def reconstruct_classification_map(q, classifier):
    l=[]
    for i in range(q.shape[0]):
      if q.iloc[i, -1] == 0:
        l.append(0)
      else:
        l.append(classifier.predict(q.iloc[i, :-1].values.reshape(1, -1)))
    return l
```

```python
clmap = np.array(l).reshape(145, 145).astype('float')
plt.figure(figsize=(10, 8))
plt.imshow(clmap, cmap='nipy_spectral')
plt.colorbar()
plt.axis('off')
plt.savefig('IP_cmap.png')
plt.show()
```
