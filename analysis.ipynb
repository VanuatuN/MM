```python
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
import numpy as np

# Data for RF
rf_accuracy = 81.03
rf_precision = [74, 63, 39, 83, 97, 99, 79, 88, 56, 97, 97, 52, 86]
rf_recall = [73, 82, 65, 95, 87, 99, 81, 75, 73, 97, 87, 81, 100]
rf_f1 = [73, 71, 49, 89, 92, 99, 80, 81, 63, 97, 92, 63, 92]

# Data for SVC
svc_accuracy = 85.49
svc_precision = [82, 75, 76, 92, 95, 99, 84, 84, 80, 98, 96, 64, 96]
svc_recall = [77, 81, 77, 89, 95, 99, 84, 84, 82, 98, 92, 82, 96]
svc_f1 = [79, 77, 77, 90, 95, 99, 84, 84, 81, 98, 94, 72, 96]

# Class labels
class_labels = [2, 3, 4, 5, 6, 8, 10, 11, 12, 13, 14, 15, 16]

# Convert raw data to classification reports
rf_classification_report = classification_report(class_labels, labels=class_labels,
                                                 target_names=[str(label) for label in class_labels],
                                                 sample_weight=None, digits=2,
                                                 output_dict=False)

svc_classification_report = classification_report(class_labels, labels=class_labels,
                                                  target_names=[str(label) for label in class_labels],
                                                  sample_weight=None, digits=2,
                                                  output_dict=False)

def plot_classification_report(report, title='Classification Report'):
    lines = report.split('\n')

    classes = []
    plotMat = []

    for line in lines[2:-3]:
        t = line.strip().split()
        if len(t) == 0:
            continue
        classes.append(t[0])
        v = [float(x) for x in t[1: len(t) - 1]]
        plotMat.append(v)

    average = lines[-1].split()
    classes.append('avg/total')
    v = [float(x) for x in average[1:len(average) - 1]]
    plotMat.append(v)

    plt.figure(figsize=(10, 8))
    plt.imshow(plotMat, interpolation='nearest', cmap=plt.cm.Blues)

    title = ' '.join(title.split('_'))
    plt.title(title)
    plt.colorbar()

    x_tick_marks = np.arange(3)
    y_tick_marks = np.arange(len(classes))

    plt.xticks(x_tick_marks, ['precision', 'recall', 'f1-score'], rotation=45)
    plt.yticks(y_tick_marks, classes)

    plt.tight_layout()
    plt.ylabel('Classes')
    plt.xlabel('Metrics')
    plt.show()

# Plot RF classification report
plot_classification_report(rf_classification_report, title='Random_Forest')

# Plot SVC classification report
plot_classification_report(svc_classification_report, title='Support_Vector_Classifier')
```
